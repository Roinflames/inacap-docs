```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array

# 1️⃣ Cargar imagen de ejemplo (puedes usar una de CIFAR-10 o tu propia imagen)
img_path = "dog.jpg"  # Cambia por la ruta de tu imagen
img = load_img(img_path, target_size=(150, 150))  # Redimensionar
x = img_to_array(img)
x = np.expand_dims(x, axis=0)

# 2️⃣ Definir generador de aumentos
datagen = ImageDataGenerator(
    rotation_range=40,      # Rotaciones aleatorias
    width_shift_range=0.2,  # Traslación horizontal
    height_shift_range=0.2, # Traslación vertical
    shear_range=0.2,        # Transformación shear
    zoom_range=0.2,         # Zoom
    horizontal_flip=True,   # Volteo horizontal
    brightness_range=[0.5, 1.5], # Ajuste de brillo
    fill_mode='nearest'     # Cómo rellenar pixeles vacíos
)

# 3️⃣ Generar imágenes aumentadas
fig, ax = plt.subplots(3, 3, figsize=(8, 8))
i = 0
for batch in datagen.flow(x, batch_size=1):
    ax[i // 3, i % 3].imshow(batch[0].astype("uint8"))
    ax[i // 3, i % 3].axis("off")
    i += 1
    if i == 9:  # Mostrar 9 imágenes
        break

plt.tight_layout()
plt.show()
```
